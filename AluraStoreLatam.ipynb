# 2. Ventas por categoría

```python
def obtener_registro_categoria(categorias_tienda, productos_tienda):
    registro_tienda = {}
    for clave, valor in zip(categorias_tienda, productos_tienda):
        if clave not in registro_tienda:
            registro_tienda[clave] = []
        registro_tienda[clave].append(valor)
    return registro_tienda

categorias_tienda1 = tienda['Categoría del Producto'].tolist()
productos_tienda1 = tienda['Producto'].tolist()
registro_tienda1 = obtener_registro_categoria(categorias_tienda1, productos_tienda1)

categorias_tienda2 = tienda2['Categoría del Producto'].tolist()
productos_tienda2 = tienda2['Producto'].tolist()
registro_tienda2 = obtener_registro_categoria(categorias_tienda2, productos_tienda2)

categorias_tienda3 = tienda3['Categoría del Producto'].tolist()
productos_tienda3 = tienda3['Producto'].tolist()
registro_tienda3 = obtener_registro_categoria(categorias_tienda3, productos_tienda3)

categorias_tienda4 = tienda4['Categoría del Producto'].tolist()
productos_tienda4 = tienda4['Producto'].tolist()
registro_tienda4 = obtener_registro_categoria(categorias_tienda4, productos_tienda4)
```

```python
def contar_productos_categoria(registro_tienda):
  contador = {}
  for categoria, productos in registro_tienda.items():
    contador[categoria] = len(productos)
  return contador

contador_tienda1 = contar_productos_categoria(registro_tienda1)
contador_tienda2 = contar_productos_categoria(registro_tienda2)
contador_tienda3 = contar_productos_categoria(registro_tienda3)
contador_tienda4 = contar_productos_categoria(registro_tienda4)
```

```python
def mostrar_grafico_pastel(conteo, nombre_tienda=''):
    # 1. Ordenar las categorías por cantidad descendente
    categorias_ordenadas = sorted(conteo.items(), key=lambda x: x[1], reverse=True)

    etiquetas = [categoria for categoria, _ in categorias_ordenadas]
    cantidades = [cantidad for _, cantidad in categorias_ordenadas]

    # 2. Crear degradado azul (más oscuro para mayor cantidad)
    cmap = cm.get_cmap('Blues')
    colores = cmap(np.linspace(0.9, 0.3, len(cantidades)))

    # 3. Función personalizada para mostrar porcentaje y cantidad
    def formato_etiqueta(pct, all_vals):
        total = sum(all_vals)
        valor_absoluto = int(round(pct / 100. * total))
        return f'{pct:.1f}%\n({valor_absoluto})'

    # 4. Crear gráfico
    plt.figure(figsize=(6, 6))
    plt.pie(
        cantidades,
        labels=etiquetas,
        colors=colores,
        autopct=lambda pct: formato_etiqueta(pct, cantidades),
        startangle=90
    )
    plt.title(f"Distribución de productos por categoría {nombre_tienda}")
    plt.axis('equal')
    plt.show()

mostrar_grafico_pastel(contador_tienda1,'Tienda 1')
mostrar_grafico_pastel(contador_tienda2,'Tienda 2')
mostrar_grafico_pastel(contador_tienda3,'Tienda 3')
mostrar_grafico_pastel(contador_tienda4,'Tienda 4')
```

# 3. Calificación promedio de la tienda


```python
def promedio_calificacion_tienda(calificaciones_tienda):
  resultado = sum(calificaciones_tienda)/len(calificaciones_tienda)
  return(resultado)

calificaciones_tienda1 = tienda['Calificación'].tolist()
calificaciones_tienda2 = tienda2['Calificación'].tolist()
calificaciones_tienda3 = tienda3['Calificación'].tolist()
calificaciones_tienda4 = tienda4['Calificación'].tolist()

promedio_calificacion_tienda1 = promedio_calificacion_tienda(calificaciones_tienda1)
promedio_calificacion_tienda2 = promedio_calificacion_tienda(calificaciones_tienda2)
promedio_calificacion_tienda3 = promedio_calificacion_tienda(calificaciones_tienda3)
promedio_calificacion_tienda4 = promedio_calificacion_tienda(calificaciones_tienda4)

print(promedio_calificacion_tienda1)
print(promedio_calificacion_tienda2)
print(promedio_calificacion_tienda3)
print(promedio_calificacion_tienda4)

promedios_calificacion = [promedio_calificacion_tienda1, promedio_calificacion_tienda2, promedio_calificacion_tienda3, promedio_calificacion_tienda4]

plt.bar(x=tiendas, height=promedios_calificacion)
plt.show
```
